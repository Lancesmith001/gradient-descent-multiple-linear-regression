```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
X1 = np.linspace(10, 20)
X2 = np.linspace(-30, 5)
Y = (17*X1) - (21*X2) + 3.5
```

```python
def predict(x1, x2, b0, b1, b2):
    return (b1*x1) + (b2*x2) + b0

def mse(ys, yh):
    return ((ys - yh) ** 2).mean()
```

```python
losses, b0s, b1s, b2s = [], [], [], []
alpha = 0.0001
B0, B1, B2 = [0.1] * 3

for i in range(100):
    yh = predict(X1, X2, B0, B1, B2)
    loss = mse(Y, yh)
    
    b0g = (-2 *  1 * (Y - yh)).mean()
    b1g = (-2 * X1 * (Y - yh)).mean()
    b2g = (-2 * X2 * (Y - yh)).mean()
    
    B0 -= (alpha * b0g)
    B1 -= (alpha * b1g)
    B2 -= (alpha * b2g)
    
    losses.append(loss)
    b0s.append(B0)
    b1s.append(B1)
    b2s.append(B2)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15,8))
ax[0, 0].plot(range(100), losses, color='#ff33cc')
ax[0, 1].plot(range(100), b0s, color='#33ffcc')
ax[1, 0].plot(range(100), b1s, color='#ffcc33')
ax[1, 1].plot(range(100), b2s, color='#33ccff')
ax[0, 0].set_xlabel('Iteration')
ax[0, 0].set_ylabel('Loss')
ax[0, 1].set_xlabel('Iteration')
ax[0, 1].set_ylabel('Beta 0')
ax[1, 0].set_xlabel('Iteration')
ax[1, 0].set_ylabel('Beta 1')
ax[1, 1].set_xlabel('Iteration')
ax[1, 1].set_ylabel('Beta 2')
plt.show()
```
